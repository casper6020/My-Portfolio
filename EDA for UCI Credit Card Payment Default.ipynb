### Relationship between the target variable and independent variables

**Observing Distribution of Customer's Age by categories of target variable**

**This distribution plot indicates that majority of credit card holders are young people.
Instead of keeping the 'Age' variable as numerical, we can transform this to different age groups, which are easy-to-interpret and explainable.**

```python
groups = ['21-30','31-40','41-50','51-60','61-70','71-80']
bins = [20,30,40,50,60,70,80]
df['age_groups'] = pd.cut(x=df['age'],bins=bins, labels=groups, include_lowest=True)
```

```python
cont_tab = pd.crosstab(df['age_groups'],df['def_pay'])
cont_tab.plot(kind='bar',stacked=True)
```

*Variables pay_0, pay_2, ...., pay_6 are also categorical. Let's visualize them used stacked bar chart.*

```python
cols = ['pay_0','pay_2','pay_3','pay_4','pay_5','pay_6']
# Plotting
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 3*2), sharey=True)
axes = axes.flatten()

for i, col in enumerate(cols):
    xdf = pd.crosstab(df[col],df['def_pay'])
    xdf.plot(kind='bar',stacked=True,ax=axes[i])
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    axes[i].get_legend().remove()
    
legend = plt.legend(title='Payment Default Status', loc='upper right', bbox_to_anchor=(1, 1), fontsize=10, borderaxespad=0)
legend.get_title().set_fontsize(10)
plt.tight_layout()
plt.show()
```

### Chi-square Test

```python
import scipy.stats as stats

cat_vars = ['marriage','education','age_groups','pay_0','pay_2','pay_3','pay_4','pay_5','pay_6']
alpha = 0.05
for cat_var in cat_vars:
    data = pd.crosstab(df[cat_var],df['def_pay'])
    chi2, p_value, dof, expected = stats.chi2_contingency(data)
    
    if p_value <= alpha:
        print(f'{cat_var} has association with payment default.')
    else:
        print(f'{cat_var} has no association with payment default.')
```

*All these variables are associated with payment default*

### ANOVA Test

```python
df.columns
```

```python
from scipy.stats import f_oneway
num_cols = ['limit_bal','bill_amt1', 'bill_amt2','bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6', 'pay_amt1',
           'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6']

for col in num_cols:
    group1 = df.loc[df['def_pay'] == 0, col]
    group2 = df.loc[df['def_pay'] == 1, col]
    f_stat, p_val = f_oneway(group1, group2)
    
    if p_val <= alpha:
        print(f'The means of {col} is different across the levels of dependent variable.')
    else:
        print(f'The means of {col} is equal across the levels of dependent variable.')
```
